```python
#adding two series in pandas

import pandas as pd
series1 = pd.Series(["ABCDEF", "DEFHGJD", "GHIDSVDR"])
series2 = pd.Series(["JKLSFDAS", "MNOAFDSA", "PQRAFEFA"])
df1 = pd.DataFrame(series1.append(series2, ignore_index = True))

df1
```

```python
3 TO add andd and delete columns and rows in pandas dataframe

data={'Maths':[67,36,78,38,56],'Science':[57,76,72,87,53],'Social':[64,68,81,50,59],'English':[47,86,98,90,59]}
df2=pd.DataFrame(data)
df2      
```

```python
# to add a new column
df2['total']=df2['Maths']+df2['Science']
df2
```

```python
# to drop a column
df2.drop(['total'],axis=1,inplace=True)
df2
```

```python
# to drop a row

df2.drop(0,axis=0,inplace=True)
df2
```

```python
# to add a new row

df2.loc[0]=[78,45,67,35]
df2
```

```python
# Explore more on set_index(), reset_index()

#set_index()
#set_index() - to change another column to an index.

df3=df2.set_index('Social')
df3
```

```python
#reset_index()
#reset_ined() - sequential numbers are set to the index and the original index is added to the data column.
df3 = df3.reset_index()
df3
```

```python
# Explore on how to combine dataframes
data2={'Spanish':[63,37,53,78,89],'Roman':[55,86,62,45,55],'Italian':[64,66,82,52,53],'English':[86,98,90,59,35]}
df5=pd.DataFrame(data2)
df5
```

```python
# Stack the DataFrames on top of each other
vertical_stack = pd.concat([df3, df5], axis=0)
vertical_stack
```

```python
# Place the DataFrames side by side
horizontal_stack = pd.concat([df3, df5], axis=1)
horizontal_stack
```

```python
#joining two dataframes

#inner join

merged_inner = pd.merge(left=df3, right=df5, left_on='English', right_on='English')
merged_inner
```

```python
#left join
merged_left = pd.merge(left=df3, right=df5, how='left', left_on='English', right_on='English')
merged_left
```

```python
#right join
merged_right = pd.merge(left=df3, right=df5, how='right', left_on='English', right_on='English')
merged_right
```
