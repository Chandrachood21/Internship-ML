```python
import seaborn as sns
%matplotlib inline
```

```python
# VIOLIN PLOT


%matplotlib inline
diamonds = sns.load_dataset('diamonds')
diamonds

sns.violinplot(x="price", y="cut", data=diamonds,palette='rainbow')
```

```python
#Strip Plot

sns.set(style = 'whitegrid') 
%matplotlib inline
flights = sns.load_dataset("flights")       
sns.stripplot(x="passengers", y="month", data=flights)
```

```python
#DIAMOND DATASET

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
dtset = sns.load_dataset("diamonds")
  
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('diamonds dataset graphs')
axes[1,2].set_axis_off()
sns.violinplot(ax=axes[0,1],x="price", y="cut", data=diamonds)
sns.stripplot(ax=axes[0,2],x="price", y="cut", data=diamonds)
sns.boxplot(ax=axes[0, 0], data=diamonds, x='price', y='cut')
sns.boxenplot(ax=axes[1,1],x="price", y="cut",scale="linear", data=diamonds)
sns.barplot(ax=axes[1, 0], data=diamonds, x='price', y='cut')


axes[0, 0].set_title("Box Plot")
axes[0, 1].set_title("Violin Plot")
axes[0, 2].set_title("Strip Plot")
axes[1, 1].set_title("Boxen Plot")
axes[1, 0].set_title("Bar Plot")
```

```python
#IRIS DATASET

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
dtset = sns.load_dataset("iris")

  
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('iris dataset graphs')
axes[1,2].set_axis_off()
sns.violinplot(ax=axes[0,1],x="species", y="petal_width", data=dtset)
sns.stripplot(ax=axes[0,2],x="species", y="petal_width", data=dtset)
sns.boxplot(ax=axes[0, 0], data=dtset, x='species', y='petal_width')
sns.barplot(ax=axes[1, 0], data=dtset, x='species', y='petal_width')
sns.boxenplot(ax=axes[1,1],x="species", y="petal_width",scale="linear", data=dtset)

axes[0, 0].set_title("Box Plot")
axes[0, 1].set_title("Violin Plot")
axes[0, 2].set_title("Stirp Plot")
axes[1, 1].set_title("Boxen Plot")
axes[1, 0].set_title("Bar Plot")
```
